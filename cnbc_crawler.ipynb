```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import json
import time
from selenium.webdriver.chrome.options import Options
from newspaper import Article
```

```python
# orgin = np.arange(0, 34180, 10)
orgin = np.arange(0, 400, 10)
```

```python
urls = []
for i in orgin:
    urls.append('https://api.queryly.com/cnbc/json.aspx?queryly_key=31a35d40a9a64ab3&query=amazon&endindex='+str(i)+'&batchsize=10&callback=&showfaceted=false&timezoneoffset=-480&facetedfields=formats&facetedkey=formats%7C&facetedvalue=!Press%20Release%7C&needtoptickers=1&additionalindexes=4cd6f71fbf22424d,937d600b0d0d4e23,3bfbe40caee7443e,626fdfcd96444f28')
```

```python
canonical_url = []
def getnews(urls):
        for j in urls:
            try:
                options = Options()
                options.add_argument("--incognito")
                # options.add_experimental_option("excludeSwitches", ["enable-automation"])
                # options.add_experimental_option('useAutomationExtension', False)
                options.chrome_exectable_path = "./crawler_final/chromedriver.exe"

                driver = webdriver.Chrome(options=options)
                driver.get(j)
                time.sleep(5)
                content = driver.find_element(By.TAG_NAME, 'Pre').text
                parsed_json = json.loads(content)
                time.sleep(8)

                for i in parsed_json['results']:
                    canonical_url.append(i["cn:liveURL"])
            except:
                continue
```

```python
getnews(urls)
```

```python
title = []
text = []
date = []
def all(url):
    for i in url:
        try:
            print(i)
            article = Article(i)
            article.download()
            article.parse()
            title.append(article.title)
            text.append(article.text)
            date.append(article.publish_date)
        except:
            continue
```

```python
all(canonical_url)
```

```python
final = pd.DataFrame({'title':title, 'content':text}, index = date)
display(final)
```
