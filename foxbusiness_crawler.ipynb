```python
import bs4
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time
from selenium.webdriver.common.keys import Keys
```

```python
num = [0, 10, 20, 30]
```

```python
t = []
c = []
x = []
def getnews(origin):
        driver = webdriver.Chrome(executable_path = 'C:/Users/user/Desktop/crawler_final/chromedriver.exe')
        driver.get("https://www.foxbusiness.com/search?q=apple&start="+origin)

        root = bs4.BeautifulSoup(driver.page_source)
        root_title = root.find_all('h3', class_ = 'title')
        root_content = root.find_all('p', class_ = 'dek')
        root_time = root.find_all('time', class_ = 'time')
        

        for i in range(len(root_title)):
            t.append(root_title[i].text)
        for i in range(len(root_content)):
            c.append(root_content[i].text)
        for i in range(len(root_time)):
            x.append(root_time[i].text)
        time.sleep(10)
        driver.close()
```

```python
for i in num:
    getnews(str(i))
```

```python
a = pd.DataFrame({'title':t, 'content':c}, index = x)
a.head(50)
```
