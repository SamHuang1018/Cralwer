```python
import random
import pandas as pd
import requests
from bs4 import BeautifulSoup
from fake_headers import Headers
from newspaper import Article
```

```python
articles_title = []
articles_text = []
articles_time = []

def get_news(stock, n_results):
    header_gen = Headers(headers=True)
    headers = header_gen.generate()

    params = {"q": stock + " stock news"}

    response = requests.get("https://news.google.com/search", headers=headers, params=params)
    soup = BeautifulSoup(response.text, "html.parser")

    url_elems = soup.find_all(class_="VDXfz")
    urls = []
    for elem in url_elems:
        urls.append("https://news.google.com" + elem["href"][1:])

    total_results = 0

    for url in urls:
        if total_results == n_results:
            break
        try:
            article = Article(url)
            article.download()
            article.parse()
            x = article.title
            y = article.text
            z = article.publish_date
            if z != None:
                articles_title.append(x)
                articles_text.append(y)
                articles_time.append(z)
                total_results += 1
        except:
            continue

    return articles_title, articles_text, articles_time
```

```python
get_news('amazon', 100)
```

```python
a = pd.DataFrame({'title':articles_title, 'content':articles_text}, index = articles_time)
a
```
